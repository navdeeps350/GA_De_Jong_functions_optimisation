First I initialize the solution using the random function and then return the encoded population using the gray code.

Then I compute the fitness by first computing the objective values of the function and then adding the absolute value of the minimum and finally dividing it with the max for fitness value so that the fitness values lie between 0 and 1.

Then I encode the probability for each individual being selected in case of the rank selection by using the formula in the slides where the index of the sorted list for the original list was caluculated by applying double argsort to the fitness values. Finally, we normalize the value by dividing it by its sum, in order to have the summation of the probabilities equal to 1.

Next, I impliment the probability for each individual being selected in case of the roulette selection, by first inverting the fitness and then normalizing it.

Next, I impliment the selection rule for selecting the pair of parents for the rank selection and the roulette selection using the probabilities received from the previous functions. 

Then I implement the crossover first by selecting the crossover point randomly for each couple and then by selecting the first part of the parent 1 and the second part of parent 2 and merging them to create the first child and then merging the first part of parent 2 and the second part of parent 1 to create the second child. So, in my case 10 couples create 20 children, i.e. each couple has 2 children.

Next, I implement the mutation by iterating through the genome of each child and then iterating throught the gene of each genome and then iterating through the bit of each gene and then selecting a random number from a uniform distribution and if the number is less than the mutation rate then I flip the bit else it stays as it is.

Next, I implement elitism. In this if the elitism is set to be true I take the parents generation and concatenate it with the next generation and then take the 20 best out of them and pass over to the next generation and if it is false I just discard the whole parents generation.

Then I implement the statistics by just by accumulating the minimum, mean and standard deviation of the fitness values and the minimum value of objective function for each iteration and accumulating in a list. 

```python
# It's a class that implements a genetic algorithm for the De Jong function
class GeneticAlgorithmForDJ:

  def __init__(self, dj_instance, selection_rule, elitism=False, population_array=None, population_size=20,
               mutation_rate=0.01, iterations=600):
    """
    This function initializes the genetic algorithm with the given parameters.

    :param dj_instance:      This is the instance of the De Jong model that you want to optimise.
    :param selection_rule:   The selection rule to be used. Can be either 'rank' or 'roulette'.
    :param elitism:          True or False. If True, the best individual in the population will be carried over
                             to the next generation. Defaults to False (optional), in this case only the population
                             is replaced by the new generation.
    :param population_array: Optional parameter, can contain an initial array with the population. It is a list of
                             lists, where each inner list correspond to an individual of the population.
    :param population_size:  The number of individuals in the population, defaults to 20 (optional).
    :param mutation_rate:    The probability of a gene mutating, defaults 0.01 (optional).
    :param iterations:       The number of iterations to run the algorithm for, defaults to 600 (optional).
    """

    self.dj_instance = dj_instance
    self.dimension = self.dj_instance.dimension
    self.dimension_bits = self.dj_instance.num_bits
    self.selection_rule = selection_rule
    self.population_size = population_size
    self.population_array = population_array
    self.mutation_rate = mutation_rate
    self.iterations = iterations
    self.elitism = elitism
    # Encoding functions
    self.encoding_function = np.vectorize(self.dj_instance.gray_encode)
    self.decoding_function = np.vectorize(self.dj_instance.gray_decode)
    self.best_solution = []
    self.min = 0
    self.iteration = 0

    # Statistics variables
    # ...
    # self.min_fitness_array = []
    # self.std_fitness_array = []
    # self.mean_fitness_array = []
    # self.min_obj_val_array = []


    # Additional variables
    # ...

  def solve(self):
    """
    This function is the main of the class `GeneticAlgorithm`. It performs the main loop of the algorithm,
    and it calls the other functions of the class.
    """
    min_fitness_array = []
    std_fitness_array = []
    mean_fitness_array = []
    min_obj_val_array = []
    # Initialise population
    # ...
    population = self.initialise_population()

    # Evaluate population
    # ...
    fitness,obj_val = self.compute_fitness(population)
    self.min = np.min(fitness)
    self.best_solution = population[np.argmin(fitness)]


    for iteration in range(self.iterations):

      # Select parents
      # ...
      selected_parents_couple = self.selection(fitness, population)

      # Generates new individuals using reproduction rules
      # ...
      next_generation = self.crossover(selected_parents_couple)

      # Form a new population
      # ...
      mutated_next_generation = self.mutation(next_generation)

      next_generation = self.reconstruction(population, mutated_next_generation)
      
      population = next_generation

      # Evaluate population
      # ...

      fitness, obj_val = self.compute_fitness(population)
      # print(population)
      if np.min(fitness) < self.min:
        self.min = np.min(fitness)
        self.best_solution = population[np.argmin(fitness)]
        self.iteration = iteration

      # print(min(self.dj_instance.evaluate(population, gray_=True)))
      
      min_fitness_1, mean_fitness_1, std_fitness_1, min_obj_value_1 = self.update_statistics(fitness, obj_val)

      # self.min_fitness_array.append(min_fitness_1)
      # self.mean_fitness_array.append(mean_fitness_1)
      # self.std_fitness_array.append(std_fitness_1)
      # self.min_obj_val_array.append(min_obj_value_1)

      min_fitness_array.append(min_fitness_1)
      mean_fitness_array.append(mean_fitness_1)
      std_fitness_array.append(std_fitness_1)
      min_obj_val_array.append(min_obj_value_1)

      

    return np.array(min_fitness_array), np.array(mean_fitness_array), np.array(std_fitness_array), np.array(min_obj_val_array)

  def initialise_population(self):
    """
    This function generates a random population of shape `population_size` and `dimension` with values between
    `range[0]` and `range[1]`.
    :return: The population encoded using gray code.
    """
    initial_numbers = np.random.uniform(low=self.dj_instance.range[0], high=self.dj_instance.range[1], size=(self.population_size, self.dj_instance.dimension))
    population = self.encoding_function(initial_numbers)
    return population

  def compute_fitness(self, population):
    """
    :return:
    """
    # population_fitness = (self.dj_instance.evaluate(population, gray_=True) + abs(min(self.dj_instance.evaluate(population, gray_=True))))/(sum(self.dj_instance.evaluate(population, gray_=True) + abs(min(self.dj_instance.evaluate(population, gray_=True)))))
    
    objective_values = self.dj_instance.evaluate(population, gray_=True)
    fitness = objective_values + abs(min(objective_values)) + 0.1
    population_fitness = fitness/max(fitness)
    # population_fitness = (self.dj_instance.evaluate(population, gray_=True) + abs(min(self.dj_instance.evaluate(population, gray_=True))))/max(self.dj_instance.evaluate(population, gray_=True) + abs(min(self.dj_instance.evaluate(population, gray_=True))))

    return population_fitness, objective_values

  def rank_selection(self, fitness):
    """
    This function compute the individual probability of selection using rank selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    i = np.argsort(np.argsort(fitness))
    # ind_prob_rank_sel = ((1/len(fitness))*(max(fitness) + (min(fitness)-max(fitness))*((fitness).argsort())/(len(fitness)-1)))
    # print(len(fitness))
    # print(max(fitness))
    # print(min(fitness) - max(fitness))
    # print(i)
    # print(len(fitness) - 1)
    prob = (1/len(fitness))*(max(fitness) + (min(fitness) - max(fitness)) * (i)/(len(fitness) - 1))
    # print(prob)
    # for i, p in enumerate(fitness.argsort()):
    #   prob_rearange[p] = prob[i] 
    ind_prob_rank_sel = np.array(prob)/sum(np.array(prob))
    # ind_prob_rank_sel = ind_prob_rank_sel/sum(ind_prob_rank_sel)
    return ind_prob_rank_sel

  def roulette_selection(self, t):
    """
    This function compute the individual probability of selection using roulette selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    # ind_prob_roul_sel = (1- ((fitness)/sum(fitness)))/sum((1- ((fitness)/sum(fitness))))
    fitness = 1/t
    ind_prob_roul_sel = fitness/np.sum(fitness)
    
    # ind_prob_roul_sel = (1 - fitness/sum(fitness))/sum(1 - fitness/sum(fitness))
    return ind_prob_roul_sel

  def selection(self, fitness, population):
    """
    This function is used to select the parents for the next generation according to a selection rule.
    """
    # Computing the probability of each individual to be selected as parent.
    # ...

    if self.selection_rule == 'rank':
      probability = self.rank_selection(fitness)
    else:
      probability = self.roulette_selection(fitness)

    selected_parents_index = []
    selected_parents_couple = []
    # print(self.selection_rule)
    # print(probability)
    
    for couple_index in range(int(self.population_size/2)):
      # index = np.random.choice(np.arange(self.population_size), size=self.population_size, p=probability, replace=True)
      selected_parents_index.append(np.random.choice(np.arange(self.population_size), size=2, p=probability, replace=False))
      # print(index)
    # selected_parents = population[index]
    selected_parents_index = np.array(selected_parents_index)

    for idx in selected_parents_index:
      selected_parents_couple.append(population[idx[0]])
      selected_parents_couple.append(population[idx[1]])
    selected_parents_couple = np.array(selected_parents_couple).reshape(int(self.population_size/2), 2, int(self.dj_instance.dimension))

    return selected_parents_couple
    # Selecting the parents for the next generation using probabilities.
    # ...

  def crossover(self, selected_parents_couple):
    """
    This function perform a single-point crossover on each pair of parents, selected by pairing the first half of
    the parents with the second half of them.
    """
    next_generation = []

    for genome in selected_parents_couple:
      parent_1 = ''.join(genome[0])
      parent_2 = ''.join(genome[1])
      crossover_point = np.random.choice(self.dimension*self.dj_instance.num_bits)
      child_1 = parent_1[:crossover_point] + parent_2[crossover_point:]
      child_2 = parent_2[:crossover_point] + parent_1[crossover_point:]
      child_1 = np.array([child_1[i:i+self.dj_instance.num_bits] for i in range(0, len(child_1), self.dj_instance.num_bits)])
      child_2 = np.array([child_2[i:i+self.dj_instance.num_bits] for i in range(0, len(child_2), self.dj_instance.num_bits)])
      next_generation.append(child_1)
      next_generation.append(child_2)

    population = np.array(next_generation)
    return population

  def mutation(self, population):
    """
    This function, decide if mutate or not each bit of each offspring in each child, by flip it, with a mutation
     rate, otherwise keep it the same.
    """
    for genome_idx, genome in enumerate(population):
      for gene_idx, gene in enumerate(genome):
        mutated_gene = ''
        for bit in gene:
          if np.random.uniform() < self.mutation_rate:
            mutated_gene += str(1 - int(bit))
          else:
            mutated_gene += bit
        population[genome_idx][gene_idx] = mutated_gene
    return population

  def elitism_selection(self, population, mutated_next_generation):
    """
    Generating a new population as combination of the old one and the new one.
    """
    elite_population = []
    
    combined_population = np.concatenate((population, mutated_next_generation), axis = 0)
    fitness_combined_population = self.compute_fitness(combined_population)
    best_fit_population = np.argsort(fitness_combined_population)
    
    for i, n in enumerate(best_fit_population):
      if i < 20:
        elite_population.append(combined_population[i])

    elite_population = np.array(elite_population)

    return elite_population

  def reconstruction(self, population, mutated_next_generation):
    """
    If elitism is enabled, this function replace the population with the 20 best individuals.
    """
    if self.elitism == 'True':
      reconstructed_elite_population = self.elitism_selection(population, mutated_next_generation)
    else:
      reconstructed_elite_population = mutated_next_generation
    
    return reconstructed_elite_population

  def update_statistics(self, fitness, obj_val):
    """
    This function can be used to compute statistics on the objective value.
    """
    min_fitness = (np.min(fitness))
    std_fitness = (np.std(fitness))
    mean_fitness = (np.mean(fitness))
    min_obj_val = (np.min(obj_val))
    
    return min_fitness, std_fitness, mean_fitness, min_obj_val

    
```

```python
# population_size = 20
selection_rules = ['rank', 'roulette']
bool_elitism = [False, True]
methods = ['rank', 'rank + elitism', 'roulette', 'roulette + elitism']
# Single-Point Crossover
# mutation_rate = 0.01
# iterations = 600
seeds = [0, 42, 666] # (Optional)

# Represent each individual of your population as a string of bits
# Each individual basically represent a point in you De Jong Function
functions = [DeJong(1), DeJong(2), DeJong(3), DeJong(5)]
```

I plot the graphs for the minimum, standard deviation and mean of the values of the fitness and the minimum values of the objective function for each method and with and without elitism and also for all the seeds and functions.

The results I obtained were good as the value of the objective function decreases with the increase in number of iterations which was the required goal. I also print the data frame to summerize the values.

```python
c_labels = ['Function', 'Method', 'Seed', 'Best solution', 'Best objective value', 'Best iteration']
data = []

for function in functions:
  print(function.function_name)
  # Plotting the results of the different methods for each seed.
  for seed in seeds:
    np.random.seed(seed)
    fig, axes = plt.subplots(nrows=1, ncols=len(methods), figsize=(36, 8))
    fig.suptitle(f'{function.function_name}, Random Seed: {seed}')

    # Creating a new instance of the GeneticAlgorithmForDJ class for each selection rule and elitism combination, running the
    # algorithm for each of the methods and plotting the results.
    k = 0
    for r_idx, sr in enumerate(selection_rules):
      for e in bool_elitism:
        GA = GeneticAlgorithmForDJ(dj_instance=function, selection_rule=sr, elitism=e)
        min_fit_list, mean_fit_list, std_fit_list, min_obj_val_list = GA.solve()
        best_solution = [function.gray_decode(i) for i in GA.best_solution]
        if e == False: 
          data.append([function.function_name, sr , seed, best_solution, min(min_obj_val_list), min_obj_val_list.argsort()[0]])
        else:
          data.append([function.function_name, sr + '+' + 'elitism' , seed, best_solution, min(min_obj_val_list), min_obj_val_list.argsort()[0]])

        # Plot
        if e == False:
          axes[k].set_title(sr)
        else:
          axes[k].set_title(sr + '+' + 'elitism' )
        axes[k].plot(min_obj_val_list, label = 'min_obj_val')
        axes[k].plot(min_fit_list, label = 'min_fit_list')
        axes[k].plot(mean_fit_list, label = 'mean_fit_list')
        axes[k].plot(std_fit_list, label = 'std_fit_list')
        
        k +=1
      
        # Populate dataframe
        # data.append([...])
    plt.legend()
  plt.show()

df = pd.DataFrame(data, columns=c_labels)
```
